## Tidy Data

### Question 5

Read in hat_observations and split `observation` into two columns, `hat_colour` and `hat_type`.

Hint: Look up the pandas documentation for `.str.split()`

```python
observation = pd.read_csv('data/hat_observations.csv').clean_names()
observation
```

```python
observation['hat_colour'] = observation.observation.str.split(pat=',').str[0]
observation['hat_type'] = observation.observation.str.split(pat=',').str[1]
observation
```

### Question 6

Unite day, month, and year columns into a column called date using a suitable separator. Then find the date where
the most berets were observed.

Hints:
- you can use the `.agg()` function to combine columns
- you may need to convert the data types of the columns you wish to combine

```python
observation['date'] = (observation["year"].astype(str) +"-"
 + observation["month"].astype(str)+"-" + observation['day'].astype(str)
)
df = observation.drop(['day','month','year','observation'],axis = 1).copy()
```

```python
(df[df.hat_type == 'beret']
.groupby('date', as_index=False)
.agg(observations = ('observation_count', 'sum'))
.sort_values('observation_count',ascending=False)
.head(1)
)
#2018-6-18 saw 9 berets
```
